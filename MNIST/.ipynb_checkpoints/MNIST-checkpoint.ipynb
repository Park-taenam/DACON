## train

```python
import numpy as np
import pandas as pd
from PIL import Image # 이미지를 불러오기 위한 라이브러리
import matplotlib.pyplot as plt # 그림을 표시하기 위한 라이브러리

%matplotlib inline
import warnings
warnings.filterwarnings("ignore")

train_data = pd.read_csv('dataset/train/train_data.csv')
train_data
```

```python
train_file_name = train_data['file_name']
train_label = train_data['label']

# image 파일을 불러와서 list에 담아줍니다.
train_images = []
for file in train_file_name:
    train_images.append(Image.open("dataset/train/"+file))
```

```python
# 랜덤으로 9개를 뽑아서 이미지를 표시해봅시다.
random_num = np.random.permutation(5000)[:9] 

plt.style.use('ggplot')
fig, ax = plt.subplots(3,3, figsize=(15,15))
fig.suptitle("Random Digits", fontsize=40)
for i in range(1,10):
    plt.subplot(3,3,i)
    idx = random_num[i-1]
    plt.imshow(train_images[idx])
plt.show()
```

```python
from collections import Counter

label_counter = Counter(train_label)

plt.figure(figsize=(10,10))
plt.title("Number Counts", fontsize=25)
plt.bar(x=list(label_counter.keys()), height =list(label_counter.values()))
plt.xticks(list(range(10)))
plt.show()

for i in range(10):
    print(f"숫자 {i} 의 개수는 {label_counter[i]}개 입니다.")
```

```python
# 이 이미지의 간단한 통계 정보를 살펴봅시다.
image = np.array(image)

print("이미지의 사이즈 : ", image.shape)
print("이미지의 최소, 최대 값 : ", image.min(), image.max())
print("이미지의 유의미한 pixel 개수 : ", (image > 0).sum())
```

```python
# 먼저 이미지 전체를 np.array 형태로 변환해줍니다.
# 또한 분석의 편의성을 위해서 flatten() 메소드를 사용해서 28 x 28 이미지를 28*28=748 array로 쭈욱 펴줄게요.
image2number = np.array([np.array(image).flatten() for image in train_image])

print("이미지의 최소, 최대 값 : ", image2number.min(), image2number.max())
print("이미지의 유의미한 pixel 개수 평균 : ", (image2number > 0).sum()/len(image2number))
```

```python
# 0 부터 9 까지의 key 값을 가지고 value로 이미지를 가진 dictionary를 구성합니다.
# 숫자들별로 이미지 분류
image_dict = {}

for i in range(10):
    idx = train_label[train_label==i].index
    image_dict[i] = image2number[idx]
```

```python
# 이제 이 데이터로 사람들이 "평균적으로" 어떻게 숫자를 쓰는지
# 각각 숫자 이미지들을 모두 더해준 다음 시각화
plt.figure(figsize=(15,8))
for i in range(10):
    plt.subplot(2,5,i+1)
    plt.grid(False)
    plt.title("Label : "+str(i), fontsize=25)
    plt.imshow(image_dict[i].sum(axis=0).reshape((28,28)))
plt.show()
```

## more complicated visualization
- 데이터셋을 전체적으로 봤을 때, 각 숫자들만의 특성이 있는지
- 차원 축소(Dimensionality Reduction) 방법 중 하나인 T-SNE 를 사용

```python
# sklearn 라이브러리를 사용합니다.
from sklearn.manifold import TSNE

# 3~5분 정도 걸려요!
image_tsne = TSNE(n_components=2).fit_transform(image2number)
```

- 모든 이미지들을 한번에 보기 쉽도록 그들의 특성만 뽑아서 그래프 그림
- 그림을 보시면 숫자 0, 1, 2, 6, 7 은 자기들끼리 잘 뭉쳐있는 모습
- 반면, 숫자 3, 4, 5, 8, 9 는 서로 섞여 있어서 쉽게 구분하기 힘들 것 같다는 생각

```python
# 예쁜 그림을 그리기 위해 seaborn 라이브러리를 사용합니다.
import seaborn as sns

plt.figure(figsize=(16,10))
plt.title("Digits T-SNE visualization", fontsize=25)
sns.scatterplot(
    x=image_tsne[:,0], y=image_tsne[:,1],
    hue=train_label,
    palette=sns.color_palette("hls", 10),
    legend="full"
)
plt.show()
```
